```python
import os
import pandas as pd
from google.cloud import bigquery

# 1. Le indicamos a la librería dónde está tu archivo de llaves locales
os.environ["GOOGLE_APPLICATION_CREDENTIALS"] = "credenciales.json"

# 2. Inicializamos el cliente oficial de BigQuery
client = bigquery.Client()

# 3. Escribimos la query (usamos el filtro de MAX fecha para no traer duplicados)
query = """
    SELECT 
        sku,
        producto,
        stock,
        precio_unitario,
        (stock * precio_unitario) AS valor_inventario,
        fecha_proceso
    FROM `mi-proyecto-etl-498923.dataset_etl.reporte_productos`
    WHERE fecha_proceso = (SELECT MAX(fecha_proceso) FROM `mi-proyecto-etl-498923.dataset_etl.reporte_productos`)
"""

# 4. Traemos la data de GCP directo a un DataFrame de Pandas local
df = client.query(query).to_dataframe()

# 5. Desplegamos las primeras filas para confirmar que funcionó
df.head()
```
